```python
# 문자열은 어떻게 만들고 사용할까? 
# 1. 큰따옴표로 사용하기

print("Hello world")

# 2. 작은따옴표로 사용하기

print('Python is fun')

# 3. 큰따옴표 3개를 사용하기
print("""Life is too short, 
      you need pyhton""")

# 4. 작은따옴표 3개를 사용하기
print('''Life is too short, you need pyhton''')
```

```python
# 문자열 연산
## 문자열 더해서 연결하기

a = "Python"
b = " is fun"

print(a + b)

## 문자열 곱하기
a = "Python"
print(a * 3)
```

```python
# 문자열 길이 구하기
a = "life is too short"
print(len(a))
```

```python
# 문자열 인덱싱과 슬라이싱

a = "Life is too short, You need Python"
print(a[0])
print(a[5])
print(a[8])
print(a[-1])
print(a[-2])
```

```python
# 문자열 슬라이싱
a = "Life is too short, You need Python"
a[0:4]
print(a[8:11])
print(a[8:11])
print(a[28:34])

print(a[ :4])
print(a[ 28:])
print(a[-6:])
print(a[:])
```
